Lets add the genre of the games to the DataFrame. The popular games. Rest will be filled as 'Other'

```python
genre = {'League of Legends':'MOBA', 'Dota 2':'MOBA', 'Heroes of Storm':'MOBA',
        'Counter-Strike: Global Offensive':'FPS', 'DayZ':'Survival', 'Survival_Evolved':'Survival'}

stream['genre'] = stream['game'].map(genre)
stream['genre'].fillna('Other', inplace=True)
stream.head(10)
```

# Visualization 

### 1) Making a Bar Graph of Popular Games

```python
# Find the number of people streaming a particular game
viewers = stream['game'].value_counts()

# We will only show the top 10 most streamed game and add the rest of the streams to other games
others_sum = viewers[10:].sum()
viewers = viewers[:10]
other_sum = pd.Series(others_sum, index=['Other Games'])
viewers = viewers.append(other_sum)

# Sort the values in descending order
viewers.sort_values(ascending=False, inplace=True)
```

### Plotting using matplotlib

```python
plt.bar(range(len(viewers)), viewers) # Plotting the skeleton bar plot. You can add a color pallette here using color parameter
plt.title('Most Popular Game Viewers') # Add a title to the plot
plt.xlabel('Games') # Setting the X and Y axis labels
plt.ylabel('Viewers')
ax = plt.subplot()  # To get the axes object
ax.set_xticks(range(len(viewers)))  
ax.set_xticklabels(viewers.index, rotation=90) # To add the x-axis labels
plt.show()
```

### Plotting using Seaborn

```python
# Plotting the same bar plot using seaborn 
ax1 = sns.barplot(x=viewers.index, y=viewers.values, palette="coolwarm")
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Most Popular Game Viewers') # Add a title to the plot
ax1.set_xlabel('Games') # Setting the X and Y axis labels
ax1.set_ylabel('Viewers')

plt.show()
```

### 2) Making a Pie Chart of Stream Viewers Location

```python
# We have the data in viewer_locations_LoL 

# We will only show the top 7 streaming locations
others_sum = viewer_locations_LoL[7:].sum()
viewer_locations_LoL = viewer_locations_LoL[:7]
other_sum = pd.Series(others_sum, index=['Other Locations'])
viewer_locations_LoL = viewer_locations_LoL.append(other_sum)

# Sort the values in descending order
viewer_locations_LoL.sort_values(ascending=False, inplace=True)
```

```python
viewer_locations_LoL
```

### Plotting using matplotlib

```python
explode=(0.1,0,0,0,0,0,0,0) # To make the first wedge, explode, we keep it 0.1 and rest 0.
colors = ['lightskyblue', 'gold', 'lightcoral', 'gainsboro', 'royalblue', 'lightpink', \
          'darkseagreen', 'sienna', 'khaki', 'gold', 'violet', 'yellowgreen']
plt.pie(viewer_locations_LoL, labels=viewer_locations_LoL.index, explode=explode, colors=colors, shadow=True, 
        startangle=345)
plt.title('League of Legends Viewers Whereabouts')
plt.show()
```

### 3) Plotting a Time-Series Graph

```python
stream['Hour'] = stream['time_dt'].dt.hour
viewers_per_hour = stream.groupby('Hour')['device_id'].count() 
viewers_per_hour
```

```python
plt.plot(range(24), viewers_per_hour)

# Lets add a 20% error, as people could just keep their browsers open
viewers_upper = viewers_per_hour + viewers_per_hour*0.20
viewers_lower = viewers_per_hour - viewers_per_hour*0.20
plt.fill_between(range(24), viewers_upper, viewers_lower, alpha = 0.15)

plt.xlabel('Hours')
plt.ylabel('Viewers')
plt.title('Viewers Across the Day')
ax = plt.subplot()  # To get the axes object
ax.set_xticks(range(24))  
ax.set_xticklabels(range(24)) # To add the x-axis labels


plt.show()
```
