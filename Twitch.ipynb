```python
# Some important imports 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Lets checkout which country has the most number of players playing LoL.

```python
stream[stream['game']=='League of Legends'].groupby('country')['login']\
    .count().reset_index().sort_values(by='login',ascending=False)
```

Lets see which device is used the most for streaming

```python
stream['player'].value_counts()
```

Lets add the genre of the games to the DataFrame. The popular games. Rest will be filled as 'Other'

```python
genre = {'League of Legends':'MOBA', 'Dota 2':'MOBA', 'Heroes of Storm':'MOBA',
        'Counter-Strike: Global Offensive':'FPS', 'DayZ':'Survival', 'Survival_Evolved':'Survival'}

stream['genre'] = stream['game'].map(genre)
stream['genre'].fillna('Other', inplace=True)
stream.head(10)
```

```python
type(stream.iloc[0, 0]) # First 0 corresponds to time column, second 0 corresponds to first entry
```

The time column in stream is a string object. We have to convert it to datetime for plotting the variations in streams with respect to time

```python
stream['time_dt'] = pd.to_datetime(stream['time'])
stream.head()
```

```python
stream_chat = pd.merge(stream, chat, on='device_id', how='outer')
stream_chat.head(20)
```
